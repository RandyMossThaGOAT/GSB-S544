Aesthetic: x = income, y = life expectancy, group = income level:




2. Import the q1data.csv dataset.

```python
import pandas as pd
from plotnine import *
#hi
```

```python
df = pd.read_csv("q1data.csv")

df1 = df.dropna()
df1.head()
```

3. Recreate the plot as best you can using plotnine.

```python
# I looked up how to do the custom colors
world_regions = {
    "africa": "cyan",
    "americas": "lawngreen",
    "asia": "deeppink",
    "europe": "yellow"
}

# I looked up how to scale the y-axis using logarithms and change the bubble size
(ggplot(df1,
aes(
  x = "income",
  y = "life_exp",
  size = "population",
  fill = "four_regions"
))
+ geom_point(shape = "o", color = "black", stroke = 0.25)
+ scale_fill_manual(values = world_regions)
+ scale_x_log10()
+ scale_y_continuous()
+ scale_size_area(max_size=18)
)
```

4. What other geometries could you use here besides geom_point? Provide an actual plot of another geometry being used here and why it’s good or bad for visualizing these data.

```python
# Same as above
world_regions = {
    "africa": "cyan",
    "americas": "lawngreen",
    "asia": "deeppink",
    "europe": "yellow"
}

# I looked up another plot type that would work wellt with these variables (smooth)
(ggplot(df1,
aes(
  x = "income",
  y = "life_exp",
  size = "population",
  fill = "four_regions"
))
+ geom_smooth()
+ scale_fill_manual(values = world_regions)
+ scale_x_log10()
+ scale_y_continuous()
+ scale_size_area(max_size=18)
)
```

I tried using a smooth plot instead of a point-based plot to represent the data. That being said, the key is quite unhelpful considering you cannot tell which wolrd region is which, but it does do a better job showing the general trend for the variables life expectancy and income. The area that it falls short is that it does not show the bubble size (indicating the population size) so that data is lost to us in this visualization.
